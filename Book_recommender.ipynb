```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df= pd.read_csv('books.csv', encoding='unicode_escape', error_bad_lines=False )
```

```python
df
```

```python
columns=['title','authors','publisher']
```

```python
def combine_features(data):
  features=[]
  for i in range(0,  data.shape[0]):
    features.append(data['title'][i]+ ' ' + data['authors'][i]+ ' '+ data['publisher'][i])
  return features
```

```python
df['combine_features']=combine_features(df)
df
```

```python
cm=CountVectorizer().fit_transform(df['combine_features'])
```

```python
cs=cosine_similarity(cm)
print(cs)
```

```python
Title=df['title'][1]
Title
```

```python
bookID=df[df.title==Title]['bookID'].values[0]
bookID
```

```python
scores=list(enumerate(cs[bookID]))
print(scores)
```

```python
sorted_scores=sorted(scores, key=lambda x:x[1], reverse=True)
sorted_scores=sorted_scores[1:]
```

```python
j=0
print('Most recommended books to '+Title+'are\n')
for item in sorted_scores:
  book_title=df[df.bookID == item[0]]['title'].values[0]
  print(j+1, book_title)
  j=j+1
  if j>=5:
    break
```
